```python
import pandas as pd
```

```python
vnd_time_data = pd.read_csv("./output/vnd_best_per_time.csv", index_col=0)
rvns_time_data = pd.read_csv("./output/rvns_best_per_time.csv", index_col=0)
bvns_time_data = pd.read_csv("./output/bvns_best_per_time.csv", index_col=0)
```

```python
from matplotlib import pyplot as plt

def plot_all_variants():
    f = plt.figure()
    f.set_figwidth(20)
    f.set_figheight(10)


    rvns = "Редукована метода променљивих околина"
    vnd = "Метода променљивог спуста"
    bvns = "Oсновна метода променљивих околина"

    colors = {rvns: 'tomato', bvns: "forestgreen", vnd: "steelblue"}

    for i in range(18):
        plt.plot(rvns_time_data.iloc[18:-1, i], label=rvns, color=colors[rvns])
        plt.plot(vnd_time_data.iloc[18:-1, i], label=vnd, color=colors[vnd])
        plt.plot(bvns_time_data.iloc[18:-1, i], label=bvns, color=colors[bvns])

    plt.xlabel("тест инстанца")
    plt.ylabel("време извршавања [s]")
    plt.legend(colors, prop={'size': 15})
```
